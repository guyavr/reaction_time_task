```python
import json
import random
import numpy as np
import pandas as pd
```

```python
tgtFileName='go_nogo_test5'

df = pd.read_csv(tgtFileName + ".csv")
df
```

```python
numtrials=df.trialnum.iloc[-1]
trialnum=df.trialnum.to_dict()
tgt_distance=df.tgt_dist.to_dict()
tgt_angle=df.tgt_angle.to_dict()
rotation=df.rotation.to_dict()
online_fb=df.online_fb.to_dict()
endpoint_feedback=df.endpoint_fb.to_dict()
clamped_fb=df.clamped_fb.to_dict()
tone=df.tone.to_dict()
frame=df.frame.to_dict()
wait_time=df.wait_time.to_dict()
between_blocks=df.between_blocks.to_dict()
blocknum=df.blocknum.to_dict()
```

```python
jsonData = {}
jsonData["numtrials"] = int(numtrials)
jsonData["trialnum"] = trialnum
jsonData["tgt_distance"] = tgt_distance
jsonData["tgt_angle"] = tgt_angle
jsonData["rotation"] = rotation
jsonData["online_fb"] = online_fb
jsonData["endpoint_feedback"] = endpoint_feedback
jsonData["clamped_fb"] = clamped_fb
jsonData["tone"] = tone
jsonData["frame"] = frame
jsonData["wait_time"] = wait_time
jsonData["between_blocks"] = between_blocks
jsonData["blocknum"] = blocknum
        
for key in jsonData.keys():
    print ("key: ", key)
    print ("value: ", jsonData[key])
    print ("")

with open(tgtFileName + ".json", 'w') as outfile:
    json.dump(jsonData, outfile)
```
